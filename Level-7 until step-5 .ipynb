Combine everything:

Random distributions
3D arrays
apply_along_axis
einsum
cumulative functions

Example: simulate student scores over 3 semesters, compute running GPA, track top 5% per semester

```python
#let's create an overiew 
import numpy as np
rng=np.random.default_rng()
#We are taking 100 student, 3 semesters, 5 subject in each semester
stu=rng.normal(loc=49,scale=10,size=(100,3,5))
#just making sure the values are inclusive
stu=np.clip(stu,0,100)
print(stu)
```

```python
#total marks per semester using einsum
#Alternative: We can use sum and axis too
sem_total=np.einsum('ijk->ij',stu)
print(sem_total)
conditions=(stu<50,(stu>=50)&(stu<=60),(stu>60)&(stu<=70),(stu>70)&(stu<=80),stu>80)
values=(5,6,7,8,9)
gpa=np.select(conditions,values)
#Alternative: nested where can be used
print(gpa)
```

```python
each_gpa=np.mean(gpa,axis=2)
print(each_gpa)
running_gpa=np.cumsum(each_gpa,axis=1)/np.arange(1,4)
print(running_gpa)
```

```python
def gpa_(gpa):
    return gpa.mean()
s=np.apply_along_axis(gpa_,axis=2,arr=gpa)
print(s)
#Alternative: Here you can use regular mean and axis
```

```python
thresh=np.percentile(sem_total,95,axis=0)
print(thresh) 
for i in range(3):
    top_5_s_no=np.where(sem_total[:,i]>=thresh[i])
    top_5=sem_total[sem_total[:,i]>=thresh[i],i]
    print(f"semester {i+1} serial numbers",top_5_s_no)
    print(f"semester {i+1} top_5",top_5)
```
